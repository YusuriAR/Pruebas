## Feature engineering

```python

#import numpy as np
import pandas as pd
from datetime import datetime
from datetime import date
import string
#from src.utils import utils
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
#from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
#from sklearn.ensemble import RandomForestClassifier
#from sklearn.model_selection import TimeSeriesSplit
#from sklearn.model_selection import GridSearchCV


from sklearn.feature_selection import VarianceThreshold

```

```python
def add_column_days(df):
    hoy = date.today().strftime("%Y-%m-%d")
    ins = df['inspection_date'].apply(lambda x: x.strftime("%Y-%m-%d"))
    fecha1 = datetime.strptime(hoy, "%Y-%m-%d")
    fecha2 = ins.apply(lambda x: datetime.strptime(x, "%Y-%m-%d"))
    df['days'] = fecha1 - fecha2
    return df

def add_column_month(df):
    df['inspection_month'] = df['inspection_date'].dt.month
    return df

def add_column_day(df):
    df['inspection_day'] = df['inspection_date'].dt.dayofweek
    return df

def add_column_violation_num(df):
    df['violations']=df['violations'].fillna("0")
    df['violation_num'] = df['violations'].apply(lambda row:[int(Numero) for Numero in row.split(".")\
                                                             if Numero.isdigit()][0])    
    return df



def add_column_pass(df):
    
    #df['results_des'] = df['results']

    dic_results = {'pass': 'pass',
                   'out of business': 'fail',
                   'no entry': 'fail',
                   'fail': 'fail',
                   'not ready': 'fail',
                   'pass w/ conditions': 'pass',
                   'business not located': 'fail'}
    df['pass'] = df.results.map(dic_results)
    return df


def add_column_risky(df):
    dic_risk = {'risk 1 (high)': 'high',
                   'risk 2 (medium)': 'medium',
                   'risk 3 (low)': 'low',
                   'all': 'high'}
    df['risky'] = df.risk.map(dic_risk)
    
    return df


def new_features(df):
    add_column_days(df)
    add_column_month(df)
    add_column_day(df)
    add_column_violation_num(df)
    add_column_pass(df)
    add_column_risky(df)
    return df


def feature_generation(df):
    df.inputs = df.drop(['aka_name', 'address', 'city', 'state','latitude', 'longitude', \
                         'location'], axis=1)
    return df.inputs

def feature_selection(df):
        #Dividir la base en train y test
    df_train,df_test = train_test_split(df,test_size=0.2,random_state=4,shuffle=True)
    
```
